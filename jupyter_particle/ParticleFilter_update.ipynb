### Perguntas 5 à 9.

```python
plt.ioff() # Desliga o modo interativo, para nao aparecerem muitas imagens no meio

frames = 1

for delta in movimentos:
    for i in range(len(pose)):
        pose[i]+=delta[i]       
        
    all_products = []
    normal = []
    # Simula a leitura do lidar
    leituras, lidar_map = inspercles.nb_simulate_lidar_fast(pose, angles, np_image)
    for elemento in particulas:
        std_x = random.normal(0,0.02)
        std_y = random.normal(0,0.02)
        std_theta = random.normal(0,0.05)

        desloc_x = delta[0] 
        desloc_y = delta[1]
        desloc_theta = delta[2] 

        elemento.x += desloc_x 
        elemento.y += desloc_y 
        elemento.theta += desloc_theta
        
        produto = mp.mpf(1)
        
        leit_teorica = inspercles.nb_simulate_lidar_fast([elemento.x, elemento.y, elemento.theta], angles, np_image, retorno = retorno_lidar_robo, output_image = False)[0]
        for i in range(len(leituras)):
            #valor = mp.mpf(ss.norm.pdf((leituras[angles[i]]), loc = (leit_teorica[angles[i]]), scale = 5))
            valor  = mp.mpf(math.exp((-(leituras[angles[i]] - leit_teorica[angles[i]]))/(2*(5**2))))
            produto = mp.mpf(produto*valor)
        all_products.append(produto)
        
    
        
    soma = mp.mpf(sum(all_products))
    normal = all_products
    alfa = mp.mpf(1)/soma
    for i in range(len(all_products)):
        all_products[i] = float(all_products[i]*alfa )
        
    novas_particulas = Particle.draw_random_sample(particulas, all_products, len(all_products)) #n numero de particulas
    particulas_pesos_depois = [p.w for p in novas_particulas]
    novas_particulas_copy = novas_particulas
    
    
    for i in range (len(novas_particulas)):
        std_x = random.normal(0,5)
        std_y = random.normal(0,5)
        std_theta = random.normal(0,0.05)

        novas_particulas[i].x += std_x
        novas_particulas[i].y += std_y
        novas_particulas[i].theta += std_theta

    pesos_total = sum(particulas_pesos_depois)
    
    for np in novas_particulas:
        np.normalize(pesos_total)
    pesos_novos = [p.w for p in novas_particulas]
    # Desenha as particulas
    ax = inspercles.nb_draw_map(color_image, pose=pose, robot=True, particles=particulas)
    particulas = novas_particulas
    #ax.imshow(occupancy_image, alph.2)
    # Desenha o mapa do lidar
    ax.imshow(lidar_map, alpha=0.5)

    plt.savefig("EeR%04d.png"%frames, bounds="tight")

    frames+=1
    plt.close('all')


plt.ion()

```
